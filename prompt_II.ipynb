```python
#Remove new cars from the dataset since the dealer sells used cars.
df = df.query('condition != "new"')
df.shape
```

```python
#Identify the columns with missing values
listColumns = df.columns.values.tolist()
ratioMissing = 0
listColumnMissing =[]
listRatioMissing =[]
listCountMissing = []
listType = []

for i in listColumns:
    ratioMissing = np.round(df[i].isna().sum()/df[i].shape[0]*100,2)
    if (ratioMissing > 0):
        listColumnMissing.append(i)
        listRatioMissing.append(ratioMissing)
        listCountMissing.append(df[i].isna().sum())
        listType.append(df[i].dtypes)
    ratioMissing = 0

ratioMissingTuple = zip(listColumnMissing, listType, listCountMissing, listRatioMissing)
ratioMissingDf = pd.DataFrame(ratioMissingTuple, columns=['Variable','Type of Variable','# Missing','% Missing'])
ratioMissingDf
```

```python
#Analyze the unique values for categorical columns with missing values and decide how to deal with the columns with missing values
ratioMissingDfCategorical = ratioMissingDf[ratioMissingDf['Type of Variable'] == 'object']

ratioMissingDfCategoricalList = ratioMissingDfCategorical['Variable'].tolist()

for i in ratioMissingDfCategoricalList:
    print(i, df[i].unique())
```

###### Actions taken for columns with missing values
- year and odometer -> Drop the records with missing values since 'year' and 'odometer' may be strong predictors of the price and do not want to impute the value for less than 1.5% of the records.

- manufactor, model, condition, cylinders, fuel, title_status, transmission, drive, size, type and paint_color -> Replace missing values with 'missing' category. These are categorical variables. They will be hot encoded for the regression model. 'missing' will have its own category.

```python
#Drop records with 'year' and 'odometer' missing
df2 = df.dropna(subset = ['year','odometer'])

#Replace the other categorical variables with missing values with 'missing'
for i in listColumnMissing:
    df2[i] = df2[i].fillna("missing")
    
df2.shape
```

```python
#Confirm there are no columns with missing values
df2.isna().sum()
```

### Analyze continous variables "odometer" and "year"

```python
#Get the distribution of the odometer to detect any data entry issues
odometerBinning = [0,1,10,100,1000,999_999,float("inf")]
df2["odometer_bin"] = pd.cut(x=df2["odometer"],right=False,bins = odometerBinning)
df2['odometer_bin'].value_counts().sort_index()
df2.shape
```

```python
#Keep records with odometer > 0 and <1,000,000. Odometers will not go beyond 999,999. It is very unlikely that odometer is 0/

df2 = df2[(df2['odometer']>0) & (df2['odometer']<1_000_000)]
df2.shape
```

```python
#Get the distribution of the year and decide what to do with old cars
yearBinning = [1900,1910,1920,1930,1940,1950,1960,1970,1980,1990,2000,2020,2022,float("inf")]
df2["year_bin"] = pd.cut(x=df2['year'],right=False,bins = yearBinning)
df2['year_bin'].value_counts().sort_index()
df2.shape
```

```python
#Keep cars manufactured after 1990.The analysis will focus on cars that are almost 32 years old 

df2 = df2[(df2['year']>=1990)]
df2.shape
```

```python
#Drop newly generated binned columns
df2 = df2.drop(columns = ['odometer_bin','year_bin'])
```

This study had the following limitations that should be revisited in the future:
- There are a large number of records with missing and other values. Instead of imputing these values, we decided to use hot encoding after assigning missing and other to different categories, which increased the dimensionality of the dataset. We can repeat our steps by trying label encoder for the variables with ordinal values. For example, the variable 'cyclinders' has values 3-12 cyclinders, other and missing. If we didn't have other and missing, we could label encode 3 cylinders with 3, 4 cylinders with 4, etc.
- The dataset could be enhanced further by pulling information from public resources. For example, features such as gas mileage, number of doors, luxury/economy indicator might also be good predictors of resale value. This information could be pulled online by the using the column for 'model'.
- The interaction between the variables should be revisited. The model with 2-degree polynomial features provided an R-sq of 72% on the training dataset. After reprocessing the dataset (use label encoder instead of one hot encoding, remove other/missing categories), polynomial features could be tested.
- Building a set of segmented models instead of one model could also improve the prediction quality. For example, the depreciation of the luxury cars might be different. We can build a separate model for luxury cars and another model for economy cars. We can also build a different model for trucks.
